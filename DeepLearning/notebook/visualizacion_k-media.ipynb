```python
# Importar libreras necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA
from sklearn.datasets import make_blobs

# Crear un conjunto de datos sinttico con seales mezcladas
X, _ = make_blobs(n_samples=1000, centers=2, n_features=2, random_state=42)
X = np.dot(X, np.random.RandomState(0).randn(2, 2))

# Aplicar ICA para separar las seales mezcladas
ica = FastICA(n_components=2)
X_ica = ica.fit_transform(X)

# Visualizacin de las seales originales y separadas
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c='blue')
plt.title("Seales Mezcladas")

plt.subplot(1, 2, 2)
plt.scatter(X_ica[:, 0], X_ica[:, 1], c='red')
plt.title("Seales Separadas por ICA")
plt.show()
```

```python
# Importar la librera necesaria
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Aplicar k-medias con 4 clusters
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Visualizar los clusters resultantes
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red')
plt.title("Clusters Formados por k-medias")
plt.show()
```
